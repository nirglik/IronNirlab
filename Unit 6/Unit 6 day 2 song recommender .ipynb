```python
response=requests.get(url)
```

```python
#creating lists for the songs and artists
songs = []
artists = []
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# define the number of iterations of our for loop 
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select(".c-title.a-no-trucate"))
num_iter
```

```python
# iterate through the result set and retrive all the data
for i in range(num_iter):
    songs.append(soup.select(".c-title.a-no-trucate")[i].get_text(strip=True)) ## getting name of title
    artists.append(soup.select(".c-label.a-no-trucate")[i].get_text(strip=True)) ## getting name of artist
```

```python
# each list becomes a column
Hot100 = pd.DataFrame({"songs":songs,
                       "artists":artists
                      })

Hot100.head(100)
```
